```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

#### Q1 load the following image and make red rectangle around face and write below rectangle “Cat” with yellow color.


```python
img=cv2.imread("original.png")
plt.imshow(img[:,:,::-1])
img=cv2.resize(img,None,fx=0.2,fy=0.2)
cv2.rectangle(img,(0,0),(img.shape[1],img.shape[0]),(0,0,255),3)
cv2.putText(img,"Cat",(0,img.shape[0]-4),cv2.FONT_HERSHEY_COMPLEX_SMALL,2,(0,255,255),5)
cv2.imshow("image",img)
key=cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Q2 load the following image and apply sharpen filter. “write your observation”


```python
blurredimg=cv2.imread("blurred.png")
img=cv2.imread("original.png")
print(type(blurredimg))
filterm=np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
    ])
out=cv2.filter2D(blurredimg,-1,filterm)
plt.imshow(out[:,:,::-1])
f, axes = plt.subplots(1, 2, figsize=(25,15))
axes[0].imshow(blurredimg[:,:,::-1]), axes[0].set_title('original')
axes[1].imshow(out[:,:,::-1]), axes[1].set_title('out')
```

### Observation
****

##### its clear that colors are now clearer and easier to distinguish this is done by sharping the edges of the image

#### Q3 load the following image and apply blur filter “write your observation”


```python
blurredimg=cv2.imread("sharpen.png")
print(type(blurredimg))
ker=11
filterm=np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
    ])
out=cv2.medianBlur(blurredimg,ker)
plt.imshow(out[:,:,::-1])
f, axes = plt.subplots(1, 2, figsize=(25,15))
axes[0].imshow(blurredimg[:,:,::-1]), axes[0].set_title('original')
axes[1].imshow(out[:,:,::-1]), axes[1].set_title('out')
```
