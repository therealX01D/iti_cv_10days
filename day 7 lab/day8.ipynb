```python
import cv2
from matplotlib import pyplot as plt
```

```python
cv2.data.haarcascades
```

```python
faceClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eyeClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
img=cv2.imread('image.jpg')
plt.imshow(img[:,:,::-1])
```

```python
img=cv2.imread('image.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plt.imshow(gray,cmap='gray')
faces=faceClassifier.detectMultiScale(gray,scaleFactor=1.05,minNeighbors=6,minSize=(50,50))

print(faces)
for x,y,w,h in faces: # comment x,y,w,h
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
    face_slice=gray[y:y+h,x:x+w]
    eyes=eyeClassifier.detectMultiScale(face_slice,scaleFactor=1.01,minNeighbors=7,minSize=(60,60))
    for x2,y2,w2,h2 in eyes:
        cv2.rectangle(img,(x+x2,y+y2),(x+x2+w2,y+y2+h2),(255,0,100),5)




plt.imshow(img[:,:,::-1])
```

```python
import cv2
from matplotlib import pyplot as plt

Classifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_fullbody.xml')
cv2.namedWindow("my_image")
def nothing(x):
    pass
cv2.createTrackbar('scale','my_image',1,1000,nothing)
cv2.createTrackbar('neighboors','my_image',1,10,nothing)
cv2.createTrackbar('size','my_image',1,120,nothing)

while True:
    img=cv2.imread('pedestrian01.jpg')
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    scale=cv2.getTrackbarPos('scale','my_image')/1000
    neighboors=cv2.getTrackbarPos('neighboors','my_image')
    size=cv2.getTrackbarPos('size','my_image')
    bodies=Classifier.detectMultiScale(gray,scaleFactor=1+scale,minNeighbors=neighboors,minSize=(size,size))
    for x,y,w,h in bodies: # comment x,y,w,h
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
   
    cv2.imshow('my_image',img)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```
