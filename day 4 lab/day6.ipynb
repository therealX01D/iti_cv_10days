# Resize (Scaling)

```python
img=cv2.imread("carkyo.jpg")
print(img.shape)
width=500
height=500
res_img=cv2.resize(img,(width,height))
print(res_img.shape)
plt.figure(figsize=(10,10))
plt.imshow(res_img)
```

```python
img=cv2.imread("carkyo.jpg")
print(img.shape)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
res_img=cv2.resize(img,None,fx=0.5,fy=0.5)
print(res_img.shape)
#plt.figure(figsize=(10,10))
plt.imshow(res_img)
```

# Line

```python
img=np.zeros((500,500,3),np.uint8)
cv2.line(img,(100,100),(400,300),(0,0,255),3)
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow("image")
def nothing(x):
    pass
cv2.createTrackbar('x0','image',0,499,nothing)
cv2.createTrackbar('x1','image',0,499,nothing)
cv2.createTrackbar('y0','image',0,499,nothing)
cv2.createTrackbar('y1','image',0,499,nothing)
cv2.createTrackbar('thick','image',1,10,nothing)
while True:
    img=np.zeros((500,500,3),np.uint8)
    x0=cv2.getTrackbarPos('x0','image')
    x1=cv2.getTrackbarPos('x1','image')
    y0=cv2.getTrackbarPos('y0','image')
    y1=cv2.getTrackbarPos('y1','image')
    thick_2=cv2.getTrackbarPos('thick','image')
    cv2.line(img,(x0,y0),(x1,y1),(0,0,255),thick_2)
    cv2.imshow('my_image',img)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
    
```

# Rectangle

```python
img=np.zeros((500,500,3),np.uint8)
cv2.rectangle(img,(100,50),(250,250),(0,0,255),3)
#cv2.imshow("image",img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(img)
```

# Circle

```python
img=np.zeros((500,500,3),np.uint8)
cv2.circle(img,(250,250),100,(255,0,0),2)
plt.imshow(img)
```

```python
cv2.namedWindow("image")
def nothing(x):
    pass
cv2.createTrackbar('x0','image',250,499,nothing)
cv2.createTrackbar('x1','image',350,499,nothing)
cv2.createTrackbar('y0','image',250,499,nothing)
cv2.createTrackbar('y1','image',250,499,nothing)
cv2.createTrackbar('thick','image',1,10,nothing)
while True:
    img=np.zeros((500,500,3),np.uint8)
    x0=cv2.getTrackbarPos('x0','image')
    x1=cv2.getTrackbarPos('x1','image')
    y0=cv2.getTrackbarPos('y0','image')
    y1=cv2.getTrackbarPos('y1','image')
    thick_2=cv2.getTrackbarPos('thick','image')
    cv2.line(img,(x0,y0),(x1,y1),(0,0,255),thick_2)
    cv2.circle(img,(250,250),100,(255,0,0),2)
    cv2.imshow('my_image',img)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```

# ploygon

```python
img=np.zeros((500,500,3),np.uint8)
pts=np.array(
    [[100,250],[250,300],[250,250],[50,100]
    ]

)
cv2.polylines(img,[pts],False,(255,255,0),3)
plt.imshow(img)
```
