```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

#### Q1
#### Make 100*100 rgb pixel image with white background and make diagonal of image
#### with black color.

```python
whiteimg=255*np.ones((100,100,3))
for i,row in enumerate(whiteimg):
    for j,col in enumerate(row):
        if(i==j):
            whiteimg[i,j]=[0,0,0]
while(True):
    cv2.imshow("window",whiteimg)
    key=cv2.waitKey(100)
    if key==ord('q'):
        break
cv2.destroyAllWindows()
```

#### Q2- run video stream using your laptop camera and save frames when i press key
#### “s” close camera when i press “q” ,name of images should be “img.png +{counter}”

```python
videocap=cv2.VideoCapture(0)
counter=0
while(True):
    stat,frame=videocap.read()
    if stat==True :
        cv2.imshow("my_video",frame)
        k=cv2.waitKey(60)
        print(k)
        if k==ord('s'):
            cv2.imwrite(f"img{counter+1}.png",frame)
            counter+=1
        if k==ord('q'):
            break
    else:
        cv2.destroyAllWindows()
        
        
        print("wrong_frame")
        break
videocap.release()
cv2.destroyAllWindows()
```

#### Q3
#### Make 100*100 rgb pixel image with black background and make diagonal of image
#### with (x) color such that x is came from rgb trackbar.

```python
blackimg=np.zeros((100,100,3))
cv2.namedWindow("colors")
cv2.createTrackbar('R','colors',0,255,lambda x:print(x))
cv2.createTrackbar('G','colors',0,255,lambda x:print(x))
cv2.createTrackbar('B','colors',0,255,lambda x:print(x))
while(True):
    r=cv2.getTrackbarPos('R','colors')
    g=cv2.getTrackbarPos('G','colors')
    b=cv2.getTrackbarPos('B','colors')
    for i,row in enumerate(blackimg):
        for j,col in enumerate(row):
            if(i==j):
                blackimg[i,j]=[b,g,r]
    cv2.imshow("colors",blackimg)
    key=cv2.waitKey(100)
    if key==ord('q'):
        break
cv2.destroyAllWindows()
```

#### Q4
#### create 100*100 rgb image with black background and make 1 vertical line and 1 horizontal line the color of both should be controlled spereat from 2 trackbar
#### First one “horz” and second one is “vert” 

```python
blackimg=np.zeros((100,100,3))
cv2.namedWindow("colors")
cv2.createTrackbar('R','colors',0,255,lambda x:print(x))
cv2.createTrackbar('G','colors',0,255,lambda x:print(x))
cv2.createTrackbar('B','colors',0,255,lambda x:print(x))
cv2.createTrackbar('horz','colors',0,249,lambda x:print(x))
cv2.createTrackbar('vert','colors',0,249,lambda x:print(x))
while(True):
    horz=cv2.getTrackbarPos('horz','colors')
    vert=cv2.getTrackbarPos('vert','colors')
    r=cv2.getTrackbarPos('R','colors')
    g=cv2.getTrackbarPos('G','colors')
    b=cv2.getTrackbarPos('B','colors')
    blackimg[vert,:]=[b,g,r]
    blackimg[:,horz]=[b,g,r]
    cv2.imshow("colors",blackimg)
    blackimg[vert,:]=[0,0,0]
    blackimg[:,horz]=[0,0,0]
    key=cv2.waitKey(100)
    if key==ord('q'):
        break
cv2.destroyAllWindows()
```
